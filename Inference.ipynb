```python
from transformers import Wav2Vec2FeatureExtractor, Wav2Vec2Processor, Wav2Vec2ForCTC, Wav2Vec2CTCTokenizer
import torch
import torchaudio
import librosa
from pydub import AudioSegment
```

```python
processor = Wav2Vec2Processor.from_pretrained('weights/wave2vec2-malay-processor')
```

```python
model = Wav2Vec2ForCTC.from_pretrained('weights/wave2vec2_bm/checkpoint-400/')
model.to('cuda')
```

```python
input_val = 'data/belajar_baca_berita__seismik_challenge/wavs/belajar_baca_berita__seismik_challenge_2.wav'
label = 'data/belajar_baca_berita__seismik_challenge/labels/belajar_baca_berita__seismik_challenge_0_1.txt'
```

```python
speech_array, sampling_rate = torchaudio.load(input_val)
speech_array = librosa.resample(speech_array[0].numpy(), sampling_rate, 16000)
```

```python
input_dict = processor(speech_array, return_tensors="pt", padding=True)
```

```python
logits = model(input_dict['input_values'].to('cuda')).logits
```

```python
pred_ids = torch.argmax(logits, dim=-1)[0]
```

```python
print("Prediction:")
print(processor.decode(pred_ids))

print('\nReference')
open(label).read()
```
